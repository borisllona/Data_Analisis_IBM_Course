```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np

df = pd.read_csv("imports-85.DATA", header = None) #header = None if we don't specify the parameters of each colm
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style",
          "drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type",
          "num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
          "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
```

```python
df["symboling"] = df["symboling"]+1 #Increment value of a column
df.replace("?", np.nan, inplace = True)
df.dropna(subset=["price"],axis=0,inplace=True) #removes (0 row) (1 column) of a value that has no value 
                                                #inplace modifies the df
```

df.replace(missing_value, new_value)
Good strategy is to replace by the mean (if its numerical value)
if it is categorical, replace with the most frequent

```python
#Type object to numeric
df["normalized-losses"] =  df["normalized-losses"].astype("float")
```

mean = df["normalized-losses"].mean() #We could say df["normalized-losses"].astype("float").mean() 
                                        #instead of transform whole column
df["normalized-losses"].replace(np.nan,mean)

                                                    FORMAT DATA


```python
#convert "mpg" to "L/100km"
df['city-L/100km'] = 235/df["city-mpg"]

df.rename(columns={"city_mpg":"city-L/100km"}, inplace = True)
```

                                                Data normalisation


Binning means categorize values into an specific range


np.linspace -> Creates the limit values for each range
np.linspace(2.0, 3.0, num=5)
array([2.  , 2.25, 2.5 , 2.75, 3.  ])

pd.cut -> Use cut when you need to segment and sort data values into bins

```python
df["price"] = df["price"].astype("float")
bins = np.linspace(min(df["price"]),max(df["price"]),4) 
group_names = ["Low","Medium","High"]

df["price-binnded"] = pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
print(df[["price-binnded","price"]])
```
